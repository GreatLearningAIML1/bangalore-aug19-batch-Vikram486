```python
#Q4.(1)  Do charges of people who smoke differ significantly  from the people who don't?
```

```python
sea.boxplot(x="smoker", y="charges", data=I_Data)
```

```python
Data_Smoker=I_Data[I_Data['smoker']=='yes']['charges']
```

```python
#Null-Hypothesis- Means of both smoker and non-smoker are same.
#Alternate-Hypothesis - Means of both smoker and non-smoker are not same.
```

```python
Data_Smoker.shape
```

```python
Data_Non_Smoker=I_Data[I_Data['smoker']=='no']['charges']
```

```python
Data_Non_Smoker.shape
```

```python
from scipy import stats
```

```python
stats.ttest_ind(Data_Smoker,Data_Non_Smoker)
```

```python
# Since p-value is less than 0.05, we can not reject null. So we have to accept alternate hypothesis.
# Means of both smoker and non-smoker are significantly different.
```

```python
#Ans:-Yes, There is significant difference in charges for smoker and non-smoker.
```

```python
#Q4.(2) Does bmi of males differ significantly from that of  females?
```

```python
sea.boxplot(x="sex", y="bmi", data=I_Data)
```

```python
#  Means of bmi for male and female are same.
#  Means of bmi for male and female are not same.
```

```python
Data_male=I_Data[I_Data['sex']=='male']['bmi']
```

```python
Data_male.shape
```

```python
Data_female=I_Data[I_Data['sex']=='female']['bmi']
```

```python
Data_female.shape
```

```python
stats.ttest_ind(Data_male,Data_female)
```

```python
# Since p-value is greater than 0.05. We can not reject null-hypothesis.
# Means of bmi for male and female are same.
```

```python
#Ans:-No, There is no significant difference in bmi for different gender.
```
